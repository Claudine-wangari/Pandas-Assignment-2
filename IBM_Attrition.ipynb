## Question 1:Show a breakdown of distance from home by job role and attrition.
The best way to do this is show the relationship using a contingency table.


The contingency table is computed using pd.crosstab

```python
#Using pd.crosstab to compute the contingency table to show a breakdown of distance from home by attrition and job role
data_crosstab = pd.crosstab([ibm['JobRole'], ibm['Attrition']],ibm['DistanceFromHome'], margins = False) 
data_crosstab
```

The contingency table can be translate into a bar plot for easier analysis

```python
#A visual represantion of breakdown of distance from home by job role and attrition
pd.crosstab([ibm['JobRole'], ibm['Attrition']],ibm['DistanceFromHome']).plot(kind='bar', stacked= True)
plt.title('Distance from Home with respect to Attrition and Job Role')
plt.xlabel('Attrition and Job Role')
plt.ylabel('Distance from Home')
```

```python
#A visual represantion of Distance from home and Job role
pd.crosstab(ibm['DistanceFromHome'],ibm['JobRole']).plot(kind='bar', stacked= True)
plt.title('Distance from Home with respect to Job role')
plt.xlabel('Distance from home')
plt.ylabel('Number of people per distance from a specific role')
```

```python
#A visual represantion of Distance from home and Attrition
pd.crosstab(ibm['DistanceFromHome'],ibm['Attrition']).plot(kind='bar', stacked= True)
plt.title('Distance from Home with respect to Attrition')
plt.xlabel('Distance from home')
plt.ylabel('Number of people per distance with a specific attrition')
```
