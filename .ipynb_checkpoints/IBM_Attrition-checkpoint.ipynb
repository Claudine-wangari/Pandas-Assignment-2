## Question 1:Show a breakdown of distance from home by job role and attrition.
The best way to do this is show the relationship using a contingency table.


The contingency table is computed using pd.crosstab

```python
#A visual represtation of breakdown of distance from home by job role and attrition
pd.crosstab([ibm['JobRole'], ibm['Attrition']],ibm['DistanceFromHome']).plot(kind='bar', stacked= True)
plt.title('Distance from Home with respect to Attrition and Job Role')
plt.xlabel('Attrition and Job Role')
plt.ylabel('Number of employees')
```

### b)Conclusion from the results

->Generally 
*the number of employees with no attrition was more than those with atrrition
*Those who had worked as Sales Executives with no attrition were the most, follwed by Research Scientists with no attrition and Laboratory Technicians with no attrition .
*Distance From Home and 1 and 2 generally had many employees for both attrition yes and no.
It can therefore be concluded that Distance from home is somehow affected by job role and attrition.

# Question 2: Compare average monthly income by education and attrition.

```python
#A visual represtation of Comparison of average monthly income by education and attrition
pd.crosstab([ibm['Education'],ibm['Attrition']],ibm['MonthlyIncome'].mean()).plot(kind='bar')
plt.title('Education and Atrrition with respect to average monthly income')
plt.xlabel('Education and Attrition')
plt.ylabel('Number of employees with the average monthly income in each category')
```

### b) Conclusion from analysis
->Generally 
*the number of employees with no attrition was more than those with atrrition
*Those who had an education level of between 2 and 4 with no attrition were the most with education level 3 with the highest.
It can therefore be concluded that average monthly income is somehow affected by education and attrition.
